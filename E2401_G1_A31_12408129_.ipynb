```python
import pandas as pd
data = pd.read_csv('C:/Users/gujja/OneDrive/Documents/new csv virat.csv')
print("Null values in data:", data.isnull().sum().sum())
print("Number of rows and columns:", data.shape)
print("Data types of each column:")
print(data.dtypes)
```

```python

subset_data = data[['Player_Name', 'Runs_Scored', 'Catches_Taken', 'Wickets_Taken']]
subset_data.to_csv('IPL2024_Performance.csv', index=False)
max_runs = subset_data.loc[subset_data['Runs_Scored'].idxmax()]
max_catches = subset_data.loc[subset_data['Catches_Taken'].idxmax()]
max_wickets = subset_data.loc[subset_data['Wickets_Taken'].idxmax()]

print("Player with maximum runs scored:", max_runs['Player_Name'])
print("Player with maximum catches taken:", max_catches['Player_Name'])
print("Player with maximum wickets taken:", max_wickets['Player_Name'])
```

```python
import matplotlib.pyplot as plt
player_data = data[data['Player_Name'] == 'Mujeeb Ur Rahman']
plt.figure(figsize=(10, 6))
plt.plot(player_data['Year'], player_data['Runs_Scored'], marker='o', label='Runs Scored')
plt.plot(player_data['Year'], player_data['Catches_Taken'], marker='o', label='Catches Taken')
plt.plot(player_data['Year'], player_data['Wickets_Taken'], marker='o', label='Wickets Taken')
plt.title("Mujeeb Ur Rahman's Performance Over the Years")
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
players_range = data.iloc[90:93]
comparison_data = data[data['Player_Name'].isin(players_range['Player_Name'])]
plt.figure(figsize=(10, 6))
for player_name in players_range['Player_Name']:
    player_data = comparison_data[comparison_data['Player_Name'] == player_name]
    plt.plot(player_data['Year'], player_data['Runs_Scored'], marker='o', label=f"{player_name} - Runs Scored")

plt.title("Comparison of Runs Scored Over the Years (Roll No. 31)")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(10, 6))
for player_name in players_range['Player_Name']:
    player_data = comparison_data[comparison_data['Player_Name'] == player_name]
    plt.plot(player_data['Year'], player_data['Catches_Taken'], marker='o', label=f"{player_name} - Catches Taken")

plt.title("Comparison of Catches Taken Over the Years (Roll No. 31)")
plt.xlabel('Year')
plt.ylabel('Catches Taken')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(10, 6))
for player_name in players_range['Player_Name']:
    player_data = comparison_data[comparison_data['Player_Name'] == player_name]
    plt.plot(player_data['Year'], player_data['Wickets_Taken'], marker='o', label=f"{player_name} - Wickets Taken")

plt.title("Comparison of Wickets Taken Over the Years (Roll No. 31)")
plt.xlabel('Year')
plt.ylabel('Wickets Taken')
plt.legend()
plt.grid()
plt.show()
```

```python
from matplotlib import pyplot as plt
yearly_totals = data.groupby('Year').sum()
plt.figure(figsize=(12, 6))
yearly_totals['Runs_Scored'].plot(kind='bar', color='blue')
plt.title('Year-wise Total Runs Scored in IPL')
plt.xlabel('Year')
plt.ylabel('Total Runs')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
yearly_totals['Catches_Taken'].plot(kind='bar', color='green')
plt.title('Year-wise Total Catches Taken in IPL')
plt.xlabel('Year')
plt.ylabel('Total Catches')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
yearly_totals['Wickets_Taken'].plot(kind='bar', color='red')
plt.title('Year-wise Total Wickets Taken in IPL')
plt.xlabel('Year')
plt.ylabel('Total Wickets')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
